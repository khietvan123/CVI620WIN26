4. 
- Add padding to all four sides of the image using a border of 50 pixels.
Try at least three border types. What visual differences do you see?
- How can you change the color of padding

```python
border_size = 50

# Constant border
border_constant = cv2.copyMakeBorder(
    image, border_size, border_size, border_size, border_size,
    cv2.BORDER_CONSTANT, value=[255, 0, 0]  # Blue border
)

# Reflect border
border_reflect = cv2.copyMakeBorder(
    image, border_size, border_size, border_size, border_size,
    cv2.BORDER_REFLECT
)

# Replicate border
border_replicate = cv2.copyMakeBorder(
    image, border_size, border_size, border_size, border_size,
    cv2.BORDER_REPLICATE
)

# Display results
titles = ["Constant", "Reflect", "Replicate"]
images = [border_constant, border_reflect, border_replicate]

plt.figure(figsize=(12,4))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(titles[i])
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.axis("off")

plt.show()

# Visual Difference:
# - Constant Border: Solid color
# - Reflect Border: Mirror effect
# - Replicate Border: Image is stretched at edges by repeating edge pixels
```

#### 12. Convolution
- Apply this kernel on an image.

         [1, 0, -1],
         [1, 0, -1],
         [1, 0, -1] 

- Use matplotlib to show your original and convolved images side by side.
- Describe the observed effect of the convolution operation. What features does this kernel highlight?

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (grayscale makes edge effects clearer)
img = cv2.imread("./Yae.jpg", cv2.IMREAD_GRAYSCALE)

# Kernel given in the lab
kernel = np.array([
    [ 1, 0, -1],
    [ 1, 0, -1],
    [ 1, 0, -1]
], dtype=np.float32)

# Apply convolution
convolved = cv2.filter2D(img, ddepth=-1, kernel=kernel)

# Show side by side using matplotlib
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Convolved")
plt.imshow(convolved, cmap="gray")
plt.axis("off")

plt.show()

# This kernel acts like a vertical edge detector.
# It compares the left side vs right side of a small 33 area.
# Vertical edges (where brightness changes leftright) become strong/visible.
# Flat/smooth areas become darker or unchanged.
```

#### 13. Denoising
- Add noise to your image.
- Apply median filtering to denoise an image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read grayscale image
img = cv2.imread("./Yae.jpg", cv2.IMREAD_GRAYSCALE)

# -------- Add salt & pepper noise --------
noisy = img.copy()
amount = 0.02  # 2% of pixels will be noisy

num_salt = int(amount * img.size / 2)
num_pepper = int(amount * img.size / 2)

# Salt (white pixels)
coords = (np.random.randint(0, img.shape[0], num_salt),
          np.random.randint(0, img.shape[1], num_salt))
noisy[coords] = 255

# Pepper (black pixels)
coords = (np.random.randint(0, img.shape[0], num_pepper),
          np.random.randint(0, img.shape[1], num_pepper))
noisy[coords] = 0

# -------- Median filter denoising --------
denoised = cv2.medianBlur(noisy, 5)  # kernel size must be odd: 3,5,7,...

# -------- Display results --------
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Noisy (Salt & Pepper)")
plt.imshow(noisy, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Denoised (Median Filter)")
plt.imshow(denoised, cmap="gray")
plt.axis("off")

plt.show()
```
